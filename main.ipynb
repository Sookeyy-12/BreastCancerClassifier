```python
from sklearn.datasets import load_breast_cancer
breast_cancer_data = load_breast_cancer()

print(breast_cancer_data.data[0])
print(breast_cancer_data.feature_names)
print(breast_cancer_data.target)
print(breast_cancer_data.target_names)
```

```python
from sklearn.model_selection import train_test_split
training_set, validation_set, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size=0.2, random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier

accuracies = []
for k in range(1, 100):
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(training_set, training_labels)
    model_score = knn_classifier.score(validation_set, validation_labels)
    accuracies.append(model_score)
```

```python
import matplotlib.pyplot as plt
k_list = range(1, 100)
plt.title("Breast Cancer Classifier Accuracy")
plt.xlabel("k")
plt.ylabel("Validatoin Accuracy")
plt.plot(k_list, accuracies)
```
